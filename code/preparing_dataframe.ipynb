```python
df = pd.read_csv('dataset/df.csv' , sep=',')
df.head()
```

```python
df['Time,EOA,COA,WOA,SOA'].apply(lambda x:x.split(',')[2])
```

```python
df['Time'] = df['Time,EOA,COA,WOA,SOA'].apply(lambda x:x.split(',')[0])
df['EOA'] = df['Time,EOA,COA,WOA,SOA'].apply(lambda x:x.split(',')[1])
df['COA'] = df['Time,EOA,COA,WOA,SOA'].apply(lambda x:x.split(',')[2])
df['WOA'] = df['Time,EOA,COA,WOA,SOA'].apply(lambda x:x.split(',')[3])
df['SOA'] = df['Time,EOA,COA,WOA,SOA'].apply(lambda x:x.split(',')[4])
df.head()
```

```python
df.drop(columns=['Time,EOA,COA,WOA,SOA'] , inplace=True)
df.head()
```
