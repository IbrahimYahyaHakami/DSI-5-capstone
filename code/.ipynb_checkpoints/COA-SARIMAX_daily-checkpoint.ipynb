```python
# load and shoe 1st 5 rows of data
df = pd.read_csv('dataset/full_COA_v3.csv' , index_col='Time',parse_dates=True)
df.head()
```

```python
# subset data set to only ['COA','temp','humidity','windSpeed']
df_l = df[['COA','temp','humidity','windSpeed']]
```

```python
# resample data to daily data
df_d = df_l.resample(rule = 'D').mean()
```

```python
# plot autocorrelation
plot_acf(df_d.COA , lags=20);
```

```python
# plot partial autocorrelation
plot_pacf(df_d.COA , lags=20);
```

```python
# define function for Dickey-Fuller Test 

def adf_test(series,title=''):
    """
    Pass in a time series and an optional title, returns an ADF report
    """
    print(f'Augmented Dickey-Fuller Test: {title}')
    result = adfuller(series.dropna(),autolag='AIC') # .dropna() handles differenced data
    
    labels = ['ADF test statistic','p-value','# lags used','# observations']
    out = pd.Series(result[0:4],index=labels)

    for key,val in result[4].items():
        out[f'critical value ({key})']=val
        
    print(out.to_string())          # .to_string() removes the line "dtype: float64"
    
    if result[1] <= 0.05:
        print("Strong evidence against the null hypothesis")
        print("Reject the null hypothesis")
        print("Data has no unit root and is stationary")
    else:
        print("Weak evidence against the null hypothesis")
        print("Fail to reject the null hypothesis")
        print("Data has a unit root and is non-stationary")
```

```python
# doing Dickey-Fuller Test
adf_test(df_d.COA)
```

```python
# define exogenous variables 
exog = df_d.drop(columns=['COA'])
```

```python
# Getting more features from the datetime
exog['year'] = df_d.index.map(lambda x: x.year)
exog['quarter'] = df_d.index.map(lambda x: x.quarter)
exog['month'] = df_d.index.map(lambda x: x.month)
exog['day'] = df_d.index.map(lambda x: x.day)
exog["weekday"]=df_d.index.map(lambda x: x.weekday())
exog["weekday"] = exog["weekday"].astype(str)
exog = pd.get_dummies(exog)
```

```python
#%A - Adding full weekday name
exog['weekday_name']=exog.index.strftime("%A")
```

```python
# Adding the hour and the weekday number
df_l['hour_24']=df_l.index.strftime("%H")
df_l['weekday_no']=df_l.index.map(lambda x:x.weekday())
```

```python
model = SARIMAX(train,
                                order=(1, 0, 0),
                            seasonal_order=(1, 0, 0, 365),
#                       exogenous=exog_to_train,
                                enforce_stationarity=False,
                                enforce_invertibility=False).fit()
```
